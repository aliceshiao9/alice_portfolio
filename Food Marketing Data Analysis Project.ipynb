```python
#Assign Age group
#Way1
def assign_age_group(Age):
    if 23 <= Age <= 30:
        return '23 - 30'
    elif 31 <= Age <= 40:
        return '31 - 40'
    elif 41 <= Age <= 50:
        return '41 - 50'
    elif 51 <= Age <= 60:  
        return '51 - 60'
    elif 61 <= Age <= 70: 
        return '61 - 70'    
    elif 71 <= Age <= 85:  
        return '71 - 85'  
    else:
        return 'unknown'

food['Age_Group'] = food['Age'].apply(assign_age_group) 
```

```python
#Assign Age group
#Way2
age_groups = [(23,30), (31,40), (41,50), (51,60), (61,70), (71,85)]
def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <= age_range[1]:
            return f'{age_range[0]} - {age_range[1]}'
    
    return 'unknown'

food['Age_Group'] = food['Age'].apply(assign_age_group) 
```

```python
#Assign Age group
#Way3
import pandas as pd
bins = [22, 30, 40, 50, 60, 70, 85]
labels = ['23 - 30', '31 - 40', '41 - 50', '51 - 60', '61 - 70', '71 - 85']
food['Age_Group'] = pd.cut(food['Age'], bins=bins, labels=labels)
```
